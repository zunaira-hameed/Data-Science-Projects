##Import Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

##Load dataset

```python
data = sns.load_dataset('iris')
data.head()
```

##Basic EDA

```python
# shape of dataset
print("Shape of dataset: ", data.shape)

# Data types and nulls
print("\nData types and nulls: \n", data.info())

# Summary statistics
print("\nSummary statistics: \n", data.describe())

# Check for missing values
data.isnull().sum()
```

## Basic Visualization

####**Scatter plot to analyzerelationship between variables**

```python
# Scatter plot: Petal Length vs Petal Width
sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=data)
plt.title('Scatter Plot: Petal Length vs Petal Width by Species')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend(title='Species')
plt.show()
```

```python
# Scatter plot: sepal length vs sepal width
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=data)
plt.title('Scatter Plot: Sepal Length vs Sepal Width by Species')
plt.show()
```

####**Histogram to examine data distribution**

```python
# Histogram : Petal length
sns.histplot(data['petal_length'], bins=20, kde=True, color='skyblue')
plt.title('Histogram of Petal Length')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Histogram : Petal width
sns.histplot(data['petal_width'], bins=20, kde=True, color='skyblue')
plt.title('Histogram of Petal Width')
plt.xlabel('Petal Width (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Histogram : Sepal length
sns.histplot(data['sepal_length'], bins=20, kde=True, color='skyblue')
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Histogram : Sepal Width
sns.histplot(data['sepal_width'], bins=20, kde=True, color='skyblue')
plt.title('Histogram of Sepal Width')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Frequency')
plt.show()
```

####**Box plot to detect outliers**

```python
# Box Plot: Sepal Length by Species
sns.boxplot(x='species', y='sepal_length', data=data)
plt.title('Box Plot of Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')
plt.show()
```

```python
# Box Plot: Sepal Width by Species
sns.boxplot(x='species', y='sepal_width', data=data)
plt.title('Box Plot of Sepal Width by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

```python
# Box Plot: Petal Length by Species
sns.boxplot(x='species', y='petal_length', data=data)
plt.title('Box Plot of Petal Length by Species')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.show()
```

```python
# Histogram : Petal Width
sns.boxplot(x='species', y='petal_width', data=data)
plt.title('Box Plot of Petal Width by Species')
plt.xlabel('Species')
plt.ylabel('Petal Width (cm)')
plt.show()
```
